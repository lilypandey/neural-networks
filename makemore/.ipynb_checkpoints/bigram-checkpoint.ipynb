```python
N = torch.zeros((27, 27), dtype = torch.int32)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
```

```python
itos = {i:s for s,i in stoi.items()}
```

```python
stoi
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
loglikelihood = 0.0
n = 0
for w in words:
    chs = ['.'] + list(w) + ['.']
    for chs1, chs2 in zip(chs, chs[1:]):
        ix1 = stoi[chs1]
        ix2 = stoi[chs2]
        prob = P[ix1][ix2]
        logprob = torch.log(prob)
        loglikelihood += logprob
        n += 1
        #print(f"{chs1}{chs2}:{P[ix1][ix2]:.4f}{logprob:.4f}")
nll = -loglikelihood
print(f"{loglikelihood=}")
print(f"{nll=}")
print(f"{nll/n=}")
```

```python
nlls = torch.zeros(5)
for i in range(5):
    x = xs[i].item()
    y = ys[i].item()
    print(f'bigram example {i+1}: {itos[x]}{itos[y]} (indexes {x}, {y})')
    print('input to neural network:', x)
    print('output prob from neural net:', probs[i])
    print('actual next character:', y)
    p = probs[i, y]
    print('probability assigned by net to corrct character:', p.item())
    logp = torch.log(p)
    print('log likelihood:', logp.item())
    nll = -logp
    print('negative log likelihood:', nll.item())
    nlls[i] = nll

print('average negative nll, i.e, loss = ', nlls.mean().item())
    
```

```python
#creating bigram dataset)
xs, ys = [], []
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
print(num)
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27,27),generator=g,requires_grad=True)
```

```python
for k in range(10):
    #forward pass
    xenc = F.one_hot(xs, num_classes = 27).float()
    logits = xenc @ W
    counts = logits.exp()
    prob = counts/counts.sum(1, keepdims = True)
    loss = -(prob[torch.arange(num), ys]).log().mean() + 0.01*(W**2).mean()
    print(loss.item())

    #backward pass
    W.grad = None
    loss.backward()

    #gradient descent
    W.data -= 100*W.grad
    
```

```python
g = torch.Generator().manual_seed(2147483647)
for i in range(5):
    out=[]
    ix=0
    while True:
        xenc = F.one_hot(torch.tensor([ix]), num_classes = 27).float()
        logits = xenc @ W
        counts = logits.exp()
        p = counts/counts.sum(1, keepdims = True)
        ix = torch.multinomial(p, num_samples = 1, replacement=True, generator=g).item()
        if ix==0:
            break
        out.append(itos[ix])
    print(''.join(out))
```
